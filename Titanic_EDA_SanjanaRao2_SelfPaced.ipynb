```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
neww=pd.read_excel('titanicdata1.xlsx')
neww
```

```python
neww.shape
```

```python
newship1=pd.DataFrame(neww,columns=['Pclass','Age','SibSp','Parch','Fare','male','Q','S','Survived'])
newship1
```

# Defining X and Y

```python
x=newship1.iloc[:,:-1].values
y=newship1.iloc[:,-1].values
```

```python
y.shape
```

```python
x.shape
```

# Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Building the Model

# 1) Performing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression()
Lr.fit(x_train, y_train)
```

**PREDICTIONS**

```python
y_predict=Lr.predict(x_test)
```

```python
y_test
```

```python
y_predict
```

**Model Evaluation using Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
confusion1=confusion_matrix(y_test,y_predict)
print(confusion1)
```

True Positive (TP): We correctly predicted the Actual Positive Data

True Negative (TN): We correctly predicted the Actual Negative Data

False Positive (FP): We incorrectly predicted the Actual Negative Data as Positive (Type I Error)

False Negative (FN): We incorrectly predicted the Actual Positive Data as Negative (Type II Error)

```python
TN=confusion1[0,0]
FP=confusion1[0,1]
FN=confusion1[1,0]
TP=confusion1[1,1]

print(confusion1)
print('TN:',TN)
print('FP:',FP)
print('FN:',FN)
print('TP:',TP)
```

**Confusion Matrix Evaluation Metrics**

Classification Accuracy: 

```python
from sklearn import metrics
acc=metrics.accuracy_score(y_test,y_predict)
acc1=(TN+TP)/(TN+TP+FN+FP)

print("Metrics Accuracy:",acc)
print("Calculated Accuracy:",acc1)
```

Classification Error:

```python
print("Classification Error percentage:",100*(1-acc))
```

TPR:TRUE POSITIVE RATE or RECALL.Also called sensitivity

TNR:TRUE NEGATIVE RATE or specificity

FPR:False Positive Rate

FNR:False Negative Rate

```python
TPR=100*metrics.recall_score(y_test,y_predict)
TNR=100*TN/(TN+FP)
FPR=100-TNR
FNR=100-TPR
pd.DataFrame([TPR,TNR,FPR,FNR],index=['TPR/SENSITIVITY','TNR/SPECIFICITY','FPR','FNR'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

From the above, we can infer that :

Precision is 83%

Recall is 73%

F1 score is 73%

Accuracy is 79%

**Therefore the accuracy obtained by performing Logistic Regression is 79%.**

# 2) KNN Classifier

**CREATING X AND Y**

```python
x=newship1.iloc[:,:-1].values
y=newship1.iloc[:,-1].values
```

```python
y.shape
```

```python
x.shape
```

**SPLITTING DATA INTO TRAIN AND TEST**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

**SCALING**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

**BUILDING THE MODEL**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_predict = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_test, y_predict)
```

```python
accuracy_score(y_test, y_predict)
```

```python
error = []
accuracy = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    predict_i = knn.predict(x_test)
    error.append(np.mean(predict_i != y_test))
    accuracy.append(accuracy_score(y_test, predict_i))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1,40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1,40), accuracy, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Accuracy Score K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
```

```python
y_predict1 = knn.predict(x_test)
```

```python
confusion_matrix(y_test, y_predict1)
```

```python
accuracy_score(y_test, y_predict1)
```

```python
print("Accuracy(in %): ",round(metrics.accuracy_score(y_test,y_predict1)*100,2), "%")
```

**Therefore the accuracy obtained using KNN Classification Model is 82.51%.**
